# Análise de Dados com Python

### Desafio:

Você trabalha em uma empresa de telecom e tem clientes de vários serviços diferentes, entre os principais: internet e telefone.

O problema é que, analisando o histórico dos clientes dos últimos anos, você percebeu que a empresa está com Churn de mais de 26% dos clientes.

Isso representa uma perda de milhões para a empresa.

O que a empresa precisa fazer para resolver isso?

Base de Dados: https://drive.google.com/drive/folders/1T7D0BlWkNuy_MDpUHuBG44kT80EmRYIs?usp=sharing <br>
Link Original do Kaggle: https://www.kaggle.com/radmirzosimov/telecom-users-dataset

```python
import pandas as pd
import numpy
import openpyxl

# Passo 1: Importar a base de dados 
tabela = pd.read_csv("telecom_users.csv") # Não passei o caminho pois o arquivo esta na mesma pasta do projeto

display(tabela) #Exibindo a Tabela Completa

# Passo 2: Visualizar a base de dados
    # - Entender as informações que voçê tem disponivel
    # - Decobrir as cagadas da base dados 

                       # - axis é o eixo, então axis -> 0 = linha; axis -> 1 = coluna
tabela = tabela.drop("Unnamed: 0", axis=1) #Aqui estou escluindo a coluna "Unnamed: 0"
# tabela = tabela.drop(2, axis=0) #Excluido a linha 2 
display(tabela) # Estou exibindo a tabela com a coluna excluida mesma coisa que o print

```

```python
# Passo 3: Tratamento de Dados
  # -- Tratando os valores que estao sendo reconhecidos de formas errada 
       # ou seja um tipo object quando na verdade é numero exemplo coluna "TotalGasto"
tabela["TotalGasto"] = pd.to_numeric(tabela["TotalGasto"], errors="coerce") 
         # Estou convertando o tipo da coluna TotalGasto em Numerico  
         # errors=coerce trantando os erros o coerce esta forcando o texto a virar numerico quando encontrar 
             # e o valores retornam vazio   
  
  # -- Tratando valor vazio 
     # -- Excluindo as colunas que tem TODOS(All) valores vazios
tabela = tabela.dropna(how="all", axis=1)     
     # -- Excluindo linhas que tem pelo menos um(any) valor vazio 
tabela = tabela.dropna(how="any", axis=0) 

print(tabela.info()) #.info() Exibe informações da Tabela 
```

```python
# Passo 4: Analise Inicial
print(tabela["Churn"].value_counts()) #Churn é a Coluna e value_counts esta fazendo contagem
print(tabela["Churn"].value_counts(normalize=True).map("{:.1%}".format)) 
                                #normalize=True traz os resultando em porcentagem 
                                #.map("{:.1%}".format) -- Aqui esta formatando para porcentual 
                                # depois do ponto a quantidade de casas depois da vigula   
```

```python
# %pip install plotly - Rodamos esse comando para instalar a biblioteca plotly para graficos
import plotly.express as px

# Passo 5: Analise Detalhada -Descobrir as causas do cancelamento
    # comparar cada coluna da base de dados com coluna churn(cancelamento)
    # precisar fazer dois passos criar o grafico e depois exibir o grafico

# Criando o grafico 
    # Para cada coluna da tabela quero cria um grafico de só vez 

for coluna in tabela.columns:
    grafico = px.histogram(tabela, x="Aposentado", color="Churn", text_auto=True) 
          # X é o que seria a horizontal do grafico  tabela-é a tabela variavel que foi criada
         # histogram é grafico que queremos exibir em X uma coluna em relação a nossa tabela
                      # color="Churn" é a coluna que separar em cores
                      # text_auto=True para exibi valores no grafico

    # Exibindo o Grafico
    grafico.show()
```
