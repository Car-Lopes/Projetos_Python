# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado
Referência do pyautogui: https://pyautogui.readthedocs.io/en/latest/quickstart.html

 Passo 1: Entrar no sistema da empresa (no link do drive exemplo)<br>
 Passo 2: Navegar até o local do relatorio (Entrar na pasta Exportar)<br>
 Passo 3: Exportar o relatorio (Fazer o DownLoad)<br>
 Passo 4: Calcular os indicadores (Faturamento e quantidade de Produtos)<br>
 Passo 5: Enviar um e-mail para a diretoria<br>

 pyautogui - Essa biblioteca ela controla o mouse, teclado e tela do PC.<br>
 pyerclip - Soluciona a texto com caracter especial <br>
 pandas - permite trazer base de dados para analisar no python <br>
 time - Soluciona o tempo de execução que foi determinado <br>
 pyautogui.click -> clicar<br>
 pyautogui.write -> escrever<br>
 pyautogui.press -> pressionar<br>
 pyautogui.hotkey -> atalho<br>

```python
import pyautogui
import pyperclip 
import time 
import pandas #permite trazer base de dados para analisar no python

# Passo 1: Entrar no sistema da empresa (no link do drive exemplo)
pyautogui.PAUSE = 1
pyautogui.press("win")
pyautogui.write("Google Chrome")
pyautogui.press("Enter")
time.sleep(2)
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
pyautogui.hotkey("ctrl","v")
pyautogui.press("Enter")

time.sleep(2)
 
# Passo 2: Navegar até o local do relatorio (Entrar na pasta Exportar)
pyautogui.click(x=-2614, y=443, clicks=2)
time.sleep(2)

# Passo 3: Exportar o relatorio (Fazer o DownLoad)
pyautogui.click(x=-2746, y=443, clicks=2)
time.sleep(5)
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
import pandas #tambem é possivel da apelido paras as bibliotecas
# Passo 4: Calcular os indicadores (Faturamento e quantidade de Produtos)
# Estamos indo até aonde o arquivo esta e armazenando na variavel tabela
# Sempre que passar um caminho sempre coloca na frente um r antes do caminho
tabela = pandas.read_excel(r"C:\Users\rafael.prado\Desktop\Projetos_Python\Intensivão Python\Aula 1\Vendas - Dez.xlsx")

#print ou display 
display (tabela)
             #Tabela variavel e nome da coluna extamente como esta e o que deseja nesse exemplo somar
faturamento = tabela["Valor Final"].sum()
quantidade = tabela["Quantidade"].sum()
print(faturamento)
print(quantidade)
```

### Vamos agora enviar um e-mail pelo outlook app do PC

```python
import pyautogui
import pyperclip
import time 

# Passo 5: Enviar um e-mail para a diretoria 


pyautogui.PAUSE = 2 

# - Entrar no email          
pyautogui.click(x=-2217, y=1596, clicks=1)    
time.sleep(2)      

# - clicar no botao escrever
pyautogui.click(x=-2761, y=180, clicks=1) 
time.sleep(2)

# - preencher as informações do email para envio
pyautogui.write("pythonimpressionador@gmail.com;ti5@mobensani.com.br")
pyautogui.press("tab")
time.sleep(2)

# inserindo o anexo
pyautogui.click(x=-1770, y=507, clicks=1) 
pyautogui.click(x=-1959, y=590, clicks=1) 
pyautogui.click(x=-1824, y=1385, clicks=1) 
pyautogui.click(x=-1242, y=487, clicks=1) 
pyperclip.copy(r"C:\Users\rafael.prado\Desktop\Projetos_Python\Intensivão Python\Aula 1")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("Enter")
pyautogui.click(x=-1537, y=1131, clicks=1) 
pyperclip.copy(r"Vendas - Dez.xlsx")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("Enter")

pyautogui.press("tab")
pyautogui.press("tab")

#Assunto do email
pyperclip.copy("Relatório de Vendas - Teste de Automação com Python")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab") 
pyautogui.press("tab")
time.sleep(1)
#corpo do email
#Esse formata o texto e se consegue colocar as variaveis concatenado
                 # (,) Sperador de milhar e .2f duas casas decimais f de float
texto = f""" Prezados, Bom dia!
                                        
O faturamento de ontem foi de: R$ {faturamento:,.2f}
A Quantidade em peças vendidas foi de: {quantidade:,} 

Atenciosamente, 

Carlos Lopes """    

pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")
time.sleep(2)

# - enviar email
pyautogui.hotkey("ctrl", "Enter")

#confirmacao = input("Enviar?")
```
